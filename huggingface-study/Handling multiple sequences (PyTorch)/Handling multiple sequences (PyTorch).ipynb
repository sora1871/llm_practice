```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

sequences = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]

# 1. 手動でトークン化して ID に変換
tokenized_sequences = [tokenizer.tokenize(seq) for seq in sequences]
ids = [tokenizer.convert_tokens_to_ids(tokens) for tokens in tokenized_sequences]

print("Tokenized sequences:")
for seq, tokens, id_list in zip(sequences, tokenized_sequences, ids):
    print(seq)
    print(tokens)
    print(id_list)
    print()

# 2. 個別にモデルへ入力
print("=== Individual inference ===")
for id_list in ids:
    input_ids = torch.tensor([id_list])  # バッチサイズ1
    outputs = model(input_ids)
    print(outputs.logits)

# 3. パディングしてバッチ化
padding_id = tokenizer.pad_token_id
max_len = max(len(id_list) for id_list in ids)

padded_ids = []
attention_masks = []

for id_list in ids:
    padding_length = max_len - len(id_list)
    padded = id_list + [padding_id] * padding_length
    mask = [1] * len(id_list) + [0] * padding_length

    padded_ids.append(padded)
    attention_masks.append(mask)

# テンソル化
input_ids = torch.tensor(padded_ids)
attention_mask = torch.tensor(attention_masks)

print("\n=== Batched input ===")
print("input_ids:")
print(input_ids)
print("attention_mask:")
print(attention_mask)

# 4. バッチでモデルへ入力
outputs = model(input_ids=input_ids, attention_mask=attention_mask)

print("\n=== Batched logits ===")
print(outputs.logits)
```
